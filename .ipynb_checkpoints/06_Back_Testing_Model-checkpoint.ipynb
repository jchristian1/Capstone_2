### Importing Libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

from imblearn.over_sampling import SMOTE

import logging 
import joblib
```

### Loading Dataset

```python
df = pd.read_csv('data/for_modeling_datasets/full_Financial_Data_for_modeling.csv')
df.head()
```

### Train Model (Remember to use treshold=0.66)

```python
def preprocess_data(full_df, target_col='Outperform', drop_cols=['PRICE VAR [%]', 'Class', 'Symbol']):
    """
    Preprocess the dataset by splitting into training, validation, and test sets,
    balancing classes with SMOTE, and standardizing features with StandardScaler.
    
    Parameters:
    - full_df: pd.DataFrame, the full merged dataset
    - target_col: str, the target variable name ('Outperform')
    - drop_cols: list, columns to drop due to data leakage or irrelevance
    
    Returns:
    - X_train, X_val, X_test: feature matrices for train, validation, and test sets
    - y_train, y_val, y_test: target variables for train, validation, and test sets
    - scaler: the StandardScaler object fitted on the training data
    """
    
    # Split the data by year
    train_data = full_df[full_df['Year'].isin([2014, 2015, 2016])]
    val_data_first_half = full_df[(full_df['Year'] == 2017) & (full_df.index < len(full_df) / 2)]
    val_data_second_half = full_df[(full_df['Year'] == 2017) & (full_df.index >= len(full_df) / 2)]
    test_data = full_df[full_df['Year'] == 2018]
    
    # Combine training and validation data for first half of 2017
    train_data = pd.concat([train_data, val_data_first_half], axis=0)
    
    logging.info(f"Training data size: {train_data.shape}")
    logging.info(f"Validation data size (second half 2017): {val_data_second_half.shape}")
    logging.info(f"Test data size (2018): {test_data.shape}")
    
    # Drop the columns that may cause data leakage
    X_train = train_data.drop(columns=drop_cols + [target_col])
    y_train = train_data[target_col]
    
    X_val = val_data_second_half.drop(columns=drop_cols + [target_col])
    y_val = val_data_second_half[target_col]
    
    X_test = test_data.drop(columns=drop_cols + [target_col])
    y_test = test_data[target_col]
    
    # Handle class imbalance using SMOTE on the training data
    smote = SMOTE(random_state=42)
    X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
    
    logging.info(f"After SMOTE, training data size: {X_train_smote.shape}")
    
    # Standardize the features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_smote)
    X_val_scaled = scaler.transform(X_val)
    X_test_scaled = scaler.transform(X_test)
    
    return X_train, X_val, X_test, y_train, y_val, y_test

def train_and_evaluate_model(model, X_train, y_train, X_val, y_val, threshold=0.5):
    # Train the model with best parameters
    model.fit(X_train, y_train)
    
    # Predict probabilities on validation data
    y_proba = model.predict_proba(X_val)[:, 1]  # Get probabilities for Class 1
    
    # Apply custom threshold
    y_pred = (y_proba >= threshold).astype(int)
    
    # Confusion Matrix
    cm = confusion_matrix(y_val, y_pred)
    ConfusionMatrixDisplay(cm).plot()
    plt.title(f"Confusion Matrix for {model.__class__.__name__}")
    plt.show()
    
    # Classification Report
    report = classification_report(y_val, y_pred)
    print(f"Classification Report for {model.__class__.__name__} with threshold {threshold}:\n")
    print(report)
    
    return y_pred

X_train, X_val, X_test, y_train, y_val, y_test = preprocess_data(df)
# Initialize Random Forest Classifier
rf = RandomForestClassifier(
    random_state=42,
    n_estimators=200,
    max_depth=15,
    max_features='log2',
    min_samples_split=5,
    class_weight={0: 1, 1: 10}
)

# Call the function to train and evaluate the model
y_pred_rf = train_and_evaluate_model(rf, X_train, y_train, X_val, y_val, threshold=0.5)
```

```python
# Columns to drop
drop_cols = ['PRICE VAR [%]', 'Class', 'Symbol']

# Target column
target_col = 'Outperform'

# Filter data for years 2017 and 2018
data_filtered = df[(df['Year'] == 2017) | (df['Year'] == 2018)].copy()

# Drop unnecessary columns
X_backtest = data_filtered.drop(columns=drop_cols + [target_col])

# Standardize the data
scaler = StandardScaler()
X_backtest_scaled = pd.DataFrame(scaler.fit_transform(X_backtest), columns=X_backtest.columns)

# Display the first few rows of the prepared dataset
X_backtest_scaled.head()
```

```python
# Predict the probabilities and labels for the backtest data
predict_proba = rf.predict_proba(X_backtest_scaled)[:, 1]  # Probabilities for class 1 (Outperformers)

# Apply the threshold of 0.63 to classify stocks as outperformers
label_predicted = (predict_proba >= 0.30).astype(int)

# Add new columns to the original dataset: 'Symbol', 'Year', 'Outperform', 'predict_proba', 'label_predicted'
data_filtered['predict_proba'] = predict_proba
data_filtered['label_predicted'] = label_predicted

# Create the final dataset with required columns
final_dataset = data_filtered[['Symbol', 'Year', target_col, 'label_predicted', 'predict_proba']]
```

```python
# Filter the dataset for 2017 and 2018 only
final_dataset_2017_2018 = final_dataset[final_dataset['Year'].isin([2017, 2018])]

# Merge the 'PRICE VAR [%]' column from the original data to get price variation
final_dataset_with_returns = pd.merge(final_dataset_2017_2018, df[['Symbol', 'Year', 'PRICE VAR [%]']], on=['Symbol', 'Year'], how='left')

# Function to calculate portfolio performance for each year
def calculate_portfolio_return(data, year):
    # Filter stocks that are predicted to outperform in the given year
    outperforming_stocks = data[(data['label_predicted'] == 1) & (data['Year'] == year)]
    
    # Check if there are any predicted outperformers
    if not outperforming_stocks.empty:
        # Calculate portfolio return as the average of the price variations for the predicted outperformers
        portfolio_return = outperforming_stocks['PRICE VAR [%]'].mean()
        print(f"Portfolio Return for {year}: {portfolio_return:.2f}%")
        
        # Output the list of selected stocks for the portfolio
        print(f"Stocks in the {year} portfolio:\n{outperforming_stocks[['Symbol', 'Year', 'predict_proba', 'PRICE VAR [%]']]}")
    else:
        print(f"No stocks predicted to outperform in {year}.")

# Calculate and display portfolio returns for 2017 and 2018
calculate_portfolio_return(final_dataset_with_returns, 2017)
calculate_portfolio_return(final_dataset_with_returns, 2018)

```
